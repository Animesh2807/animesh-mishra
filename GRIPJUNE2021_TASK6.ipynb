### Construction of Decision Tree Model 


```python
# 75% training data and 25% test data
# Here X_train,Y_train is training data while X_test,Y_test is testing data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(features_frame[iris.feature_names], features_frame['target'], random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
# Making an instance of the model 
clf = DecisionTreeClassifier()
# Training the model 
clf.fit(X_train, Y_train)
```

### Visualising the trained Model 

```python
from sklearn import tree
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(clf,
               feature_names = fn, 
               class_names=cn,
               filled = True);
```

### Evaluating the performance of the model 

```python
accuracy_train = clf.score(X_train,Y_train)
accuracy_test = clf.score(X_test,Y_test)
print("Train Accuracy of the decision trees is found to be: ",accuracy_train)
print("Test Accuracy of the decision trees is found to be: ",accuracy_test)
```

```python
from sklearn.metrics import precision_score
X_pred = clf.predict(X_train)
Y_pred = clf.predict(X_test)
train_micro_precision = precision_score(X_pred, Y_train, average="micro")
print('Micro-averaged precision score training data:',train_micro_precision)
test_micro_precision = precision_score(Y_pred, Y_test, average="micro")
print('Micro-averaged precision score testing data:',test_micro_precision)
```

```python
from sklearn.metrics import recall_score
train_micro_recall = recall_score(Y_train,X_pred,average="micro")
print(f"Micro-average recall score training data: {train_micro_recall}")
test_micro_recall = recall_score(Y_test,Y_pred,average="micro")
print(f"Micro-average recall score testing data: {test_micro_recall}")
```

As precision is equal to recall, there might be a probability that number of  False positive is equal to the number of False Negative
